```python
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
import tensorflow as tf
import keras as K
import pandas as pd
import os
import cv2
```

```python
train_data = pd.read_csv('../input/shopee-product-matching/train.csv')
train_data
```

```python
path = '../input/shopee-product-matching/train_images'
train_img = []
img_size = 200

def preprocess_image(image):
    url=path+'/'+image
    img_raw=tf.io.read_file(url)
    img = tf.io.decode_image(img_raw)
    img = tf.image.convert_image_dtype(img, tf.float32)
    img=tf.image.resize(img,(img_size,img_size))
    return img
        #imgs.append([np.array(img),np.eye(len(labels))[])
```

```python
train_image=[]
for f in train_data['image']:
    train_image.append(preprocess_image(f))
```
